```python
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerTuple
import seaborn as sns
import statsmodels.api as sm
```

```python
shot_data_2022 = pd.read_csv('shots_2022.csv')
time_to_shot_data = shot_data_2022[['goal', 'speedFromLastEvent', 'timeSinceLastEvent', 'teamCode', 'xGoal', 'distanceFromLastEvent', 'shotDistance']]
```

```python
teamShots = time_to_shot_data.value_counts('teamCode')
time_to_shot_data = time_to_shot_data.groupby('teamCode').sum()
time_to_shot_data['totalTeamShots'] = teamShots
time_to_shot_data['avgSpeedFromLastEvent'] = time_to_shot_data['speedFromLastEvent']/time_to_shot_data['totalTeamShots']
time_to_shot_data['avgxGoal'] = time_to_shot_data['xGoal']/time_to_shot_data['totalTeamShots']
time_to_shot_data['avgTimeSinceLastEvent'] = time_to_shot_data['timeSinceLastEvent']/time_to_shot_data['totalTeamShots']
time_to_shot_data['avgDistanceFromLastEvent'] = time_to_shot_data['distanceFromLastEvent']/time_to_shot_data['totalTeamShots']
time_to_shot_data['avgShotDistance'] = time_to_shot_data['shotDistance']/time_to_shot_data['totalTeamShots']
```

```python
team_wins = pd.read_csv('teamWins.csv')
team_wins = team_wins.rename(columns={'Unnamed: 2': 'teamCode', 'Unnamed: 1': 'Wins'})
team_wins = team_wins[['Wins', 'teamCode']]
```

```python
time_to_shot_data = time_to_shot_data.merge(team_wins, how='inner', on='teamCode')
```

Check the relationships between wins, and time from last event, speed from last event and distance from last event

```python
speedModel = sm.OLS.from_formula('xGoal ~ avgSpeedFromLastEvent', data=time_to_shot_data).fit()
print(speedModel.summary())
```

```python
timeModel = sm.OLS.from_formula('xGoal ~ avgTimeSinceLastEvent', data=time_to_shot_data)
timeModelRes = timeModel.fit()
print(timeModelRes.summary())
```

```python
distanceModel = sm.OLS.from_formula('xGoal ~ avgDistanceFromLastEvent', data=time_to_shot_data).fit()
print(distanceModel.summary())
```

```python
shotDistanceModel = sm.OLS.from_formula('xGoal ~ avgShotDistance', data=time_to_shot_data).fit()
print(shotDistanceModel.summary())
```

```python
shotsModel = sm.OLS.from_formula('goal ~ totalTeamShots', data=time_to_shot_data).fit()
print(shotsModel.summary())
```

```python
fig, axes = plt.subplots(1,4, figsize=(20,5))

fig.suptitle('Distance and Time Leading to Goals')

ax1 = sns.regplot(ax=axes[0], data=time_to_shot_data, x='avgTimeSinceLastEvent', y='xGoal' )
ax1.set_xlabel('Average Time from Last Event to Goal (sec)')

ax2 = sns.regplot(ax=axes[1], data=time_to_shot_data, x='avgSpeedFromLastEvent', y='xGoal')
ax2.set_xlabel('Average Speed from Last Event to Goal (ft/sec)')

ax3 = sns.regplot(ax=axes[2], data=time_to_shot_data, x='avgDistanceFromLastEvent', y='xGoal')
ax3.set_xlabel('Average Distance from Last Event to Goal (feet)')

ax4 = sns.regplot(ax=axes[3], data=time_to_shot_data, x='avgShotDistance', y='xGoal')
ax4.set_xlabel('Average Shot Distance per Goal (feet)')
```

```python
winsPercentile_75 = np.percentile(time_to_shot_data['Wins'], 75)
avgDistancePercentile_10 = np.percentile(time_to_shot_data['avgDistanceFromLastEvent'], 10)
distanceMean = np.mean(time_to_shot_data['avgDistanceFromLastEvent'])
goalsMean = np.mean(time_to_shot_data['goal'])
winsMean = np.mean(time_to_shot_data['Wins'])
shotsMean = np.mean(time_to_shot_data['totalTeamShots'])

pointSize = [(x-winsMean) * 10 if x > winsMean else x * 2 for x in time_to_shot_data['Wins']]
size = [x/10 for x in time_to_shot_data['totalTeamShots']]

z=[]
for i in range(0, len(time_to_shot_data['teamCode'])):
  z.append('{}'.format(time_to_shot_data['teamCode'].iat[i]))

plt.scatter(data=time_to_shot_data, x='avgDistanceFromLastEvent', y='goal', s=pointSize)
plt.axvline(distanceMean)
plt.axhline(goalsMean)

plt.title('Average Distance from an Event Immediately Leading to Shot')
plt.xlabel('Average Event Distance from Goal(feet)')
plt.ylabel('Total Goals per Team')

for i, txt in enumerate(z):
  if time_to_shot_data['Wins'].iat[i] >= winsPercentile_75:
    plt.annotate(txt, (time_to_shot_data['avgDistanceFromLastEvent'].iat[i], time_to_shot_data['goal'].iat[i]), xytext=(time_to_shot_data['avgDistanceFromLastEvent'].iat[i] +.08, time_to_shot_data['goal'].iat[i] + .08))
  

plt.show()
plt.clf()
```

```python
timeMean = np.mean(time_to_shot_data['avgTimeSinceLastEvent'])

y=[]
for i in range(0, len(time_to_shot_data['teamCode'])):
  y.append('{}'.format(time_to_shot_data['teamCode'].iat[i]))

plt.scatter(data=time_to_shot_data, x='avgTimeSinceLastEvent', y='goal', s=pointSize)
plt.axvline(timeMean)
plt.axhline(goalsMean)

plt.title('Average Time from an Event Immediately Leading to Shot')
plt.xlabel('Average Event Time Before Shot(sec)')
plt.ylabel('Total Shots per Team')

for i, txt in enumerate(y):
  if time_to_shot_data['Wins'].iat[i] >= winsPercentile_75:
    plt.annotate(txt, (time_to_shot_data['avgTimeSinceLastEvent'].iat[i], time_to_shot_data['goal'].iat[i]), xytext=(time_to_shot_data['avgTimeSinceLastEvent'].iat[i] +.08, time_to_shot_data['goal'].iat[i] + .08))
  

plt.show()
plt.clf()
```
